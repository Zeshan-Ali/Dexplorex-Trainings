```python
import numpy as np
import matplotlib.pyplot as plt

# Generate n values
n = np.arange(0, 50, 1)
f = 5 * n + 3

# Plot
plt.figure(figsize=(8, 5))
plt.plot(n, f, label='f(n) = 5n + 3', color='blue')
plt.title('Linear Growth: f(n) = 5n + 3')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Generate n values
n = np.arange(0, 50, 1)
f = n**2 + 2*n + 1

# Plot
plt.figure(figsize=(8, 5))
plt.plot(n, f, label='f(n) = n^2 + 2n + 1', color='green')
plt.title('Quadratic Growth: f(n) = n^2 + 2n + 1')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Generate n values (avoiding zero for log)
n = np.arange(1, 50, 1)
f = np.log(n)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(n, f, label='f(n) = log(n)', color='orange')
plt.title('Logarithmic Growth: f(n) = log(n)')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Generate n values
n = np.arange(0, 10, 1)  # Restrict range to prevent very large numbers
f = 2**n

# Plot
plt.figure(figsize=(8, 5))
plt.plot(n, f, label='f(n) = 2^n', color='red')
plt.title('Exponential Growth: f(n) = 2^n')
plt.xlabel('n')
plt.ylabel('f(n)')
plt.grid(True)
plt.legend()
plt.show()
```
